```python

# JamalNasir comments: BankAccount class
class BankAccount():

    # JamalNasir comments: constructor/Initializer to create account
    def __init__(self, account_number, account_holder, initial_balance=0):

        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = initial_balance
        
        # JamalNasir comments: list to hold transaction
        self.transactions = []
        
    # JamalNasir comments: deposit function taking amount
    def deposit(self, amount):

        # JamalNasir comments: check if user input the positve amount not in negetive
        # JamalNasir comments: like -500 is not a positive number
        if amount > 0:
            self.balance += amount
            self.add_transaction(f"Deposited: ${amount:.2f}")
        else:
            print("Deposit amount must be positive.")

    # JamalNasir comments: wihdraw asking amouont
    def withdraw(self, amount):

        # JamalNasir comments: check if account has enough balance in account
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                
                # JamalNasir comments: calling transaction function and passing amount
                self.add_transaction(f"Withdrew: ${amount:.2f}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    # JamalNasir comments: show balance in balance attribute
    def check_balance(self):

        return self.balance

    
    # JamalNasir comments: function to add transcation
    def add_transaction(self, description):
        
        self.transactions.append(description)

    # JamalNasir comments: function to print statement
    def print_statement(self):

        print(f"Account Statement for {self.account_holder} (Account No: {self.account_number}):")
        print("-" * 40)

        # JamalNasir comments: printing all transactioins
        # # JamalNasir comments: used loop to get all tranaction
        for transaction in self.transactions:
            print(transaction)
        print("-" * 40)
        print(f"Current Balance: ${self.balance:.2f}")



"""
if __name__ == "__main__":
    account = BankAccount("12345678", "Jamal Nasir", 1000)
    account.deposit(500)
    account.withdraw(200)
    account.withdraw(1500)  # Should print insufficient funds
    account.print_statement()
"""

class Bank:
    def __init__(self):
        
        # JamalNasir comments: list of accounts
        self.accounts = {}
        self.next_account_number = 100200100

    # JamalNasir comments: creating new account and setting 0 balance and getting name as a account holder.
    # account holder / account title / user Nasir
    
    def open_account(self, account_holder, initial_balance=0):
        account_number = self.next_account_number
        self.next_account_number += 1

        # JamalNasir comments: making object of BankAccount Class
        # new account will have bankaccount object
        # passing values to bank account class constructor/Initializer
        
        new_account = BankAccount(account_holder, account_number, initial_balance)

        # JamalNasir comments: inserting new account in account list
        self.accounts[account_number] = new_account
        return account_number

    # JamalNasir comments: get accouont function to get account from account list
    def get_account(self, account_number):
        return self.accounts.get(account_number, None)

    # JamalNasir comments: transfer function is taking send account and receiver account with amount transfer
    
    def transfer(self, sender_account_number, receiver_account_number, amount):
        sender = self.get_account(sender_account_number)
        receiver = self.get_account(receiver_account_number)

        if sender is None:
            raise ValueError(f"Sender account {sender_account_number} does not exist.")
        if receiver is None:
            raise ValueError(f"Receiver account {receiver_account_number} does not exist.")

        sender.withdraw(amount)
        receiver.deposit(amount)

    def admin_check_total_deposit(self):
        return sum(account.balance for account in self.accounts.values())

    def admin_check_total_accounts(self):
        return len(self.accounts)

# Example usage
if __name__ == "__main__":
    
    bank = Bank()

    # Open accounts
    account_1 = bank.open_account("Jamal", 1000)
    account_2 = bank.open_account("Nasir", 500)

    print(f"Account Number: {account_1} opened - Account Title: Jamal  Balance: 1000.")
    print(f"Account Number: {account_2} opened - Account Title Nasir Balance: 500.")
    print()
    
    print('Jamal transfer 200 to Nasir')
    # Transfer money
    bank.transfer(account_1, account_2, 200)
    print()
    

    print('Updated Account Balances')
    print()
    # Check balances
    print(f"Jamal's balance: {bank.get_account(account_1).balance}")
    print(f"Nasir's balance: {bank.get_account(account_2).balance}")
    print()
    print()

    # Admin checks
    print(f"Total deposits in the bank: {bank.admin_check_total_deposit()}")
    print(f"Total accounts in the bank: {bank.admin_check_total_accounts()}")

```
